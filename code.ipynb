```python
import matplotlib.pyplot as plt
import cv2

face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")        #face machine learning model and data is given by me in a excel file
eye_cascade = cv2.CascadeClassifier("haarcascade_eye.xml") 
smile_cascade=cv2.CascadeClassifier("haarcascade_smile.xml")                        #smile machine learning model and data is given by me in a excel file

img = cv2.imread("../input/nirmaa/WhatsApp Image 2022-02-13 at 5.53.54 PM.jpeg")    #test image
windows_name='image'
i=0
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

face = face_cascade.detectMultiScale(img,minNeighbors=7 , scaleFactor=1.1)     #scale in(zooming)  #co ordinates
for x,y,w,h in face:
  img = cv2.rectangle(img , (x,y) , (x+w, y+h), (0,0,255),2)                   #source image , co ordinates of box, to make rectangle , color code(red) , border thicknes
  i=i+1
  cv2.putText(img ,f'person', (x,y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,235,235), 1)
               

eye = eye_cascade.detectMultiScale(img,minNeighbors=2 , scaleFactor=1.1)   #scale in(zooming)  #co ordinates
for x,y,w,h in eye:
  img = cv2.rectangle(img , (x,y) , (x+w, y+h), (0,255,0),1)

smile = smile_cascade.detectMultiScale(gray,minNeighbors=25 , scaleFactor=1.1)   #scale in(zooming)  #co ordinates
for x,y,w,h in smile:
  img = cv2.rectangle(img , (x,y) , (x+w, y+h), (255,0,0),1)

plt.imshow(img, cmap='gray')
```
